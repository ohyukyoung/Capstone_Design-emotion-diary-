# user_item_m (1차 추천점수)

```python
#파일 불러오기

import pandas as pd
import numpy as np

exercise_level = pd.read_csv('./exercise_level.csv')
exercise_level.head()

ob_level = pd.read_csv('./ObesityDataSet_level.csv')
ob_level.head()
```

```python
# ob_level데이터에 user_id 생성
ob_level['user_id'] = ob_level.index + 1
ob_level.head()
```

```python
# excersize_level데이터에 exercise_id 생성
exercise_level['exercise_id'] = exercise_level.index + 1
exercise_level.head()
```

비만 레벨과 운동레벨 간의 차이를 이용하여 1차 추천점수를 계산하고자 함

```python

# 가정: user_data와 exercise_data는 이미 정의된 데이터프레임
# user_data: ['UserID', 'Weight', 'CAEC', 'Age', 'FAVC', 'Height', 'level']
# exercise_data: ['ExerciseID', 'Calories_per_kg', 'level']

# 사용자 수와 운동 종류 수에 따른 0으로 채워진 행렬 생성
user_item_matrix = pd.DataFrame(np.zeros((ob_level.shape[0], exercise_level.shape[0])), 
                                index=ob_level['user_id'], 
                                columns=exercise_level['exercise_id'])

for i, user_row in ob_level.iterrows():
    for j, exercise_row in exercise_level.iterrows():
        # 비만 레벨과 운동 레벨 간의 차이에 기반한 점수 계산
        score = 4 - abs(user_row['level'] - exercise_row['level'])
        # 해당 사용자-운동 쌍에 점수 할당
        user_item_matrix.loc[user_row['user_id'], exercise_row['exercise_id']] = score

print(user_item_matrix)
```
